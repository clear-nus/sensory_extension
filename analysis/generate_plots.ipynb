```python
df_all = pd.read_excel('question5.ods', engine='odf')
df_all = df_all.fillna(0)

for algorithm in ['svmlinear', 'svmrbf', 'mlp', 'rnn']:

    for feature in ['baseline', 'fft', 'kernel']:

        fig, ax = plt.subplots(figsize=(6, 3))
        df = df_all[(df_all.algorithm == algorithm) & (df_all.feature == feature)]

        for tool_length in [20, 30, 50]:

            if feature != 'kernel': biotac_means = df[['frequency', 'mean']][(df.signal == 'biotac') & (df.tool == tool_length)]
            neutouch_means = df[['frequency', 'mean']][(df.signal == 'neutouch') & (df.tool == tool_length)]

            if feature != 'kernel': biotac_std = df['std'][df.signal == 'biotac']
            neutouch_std = df['std'][df.signal == 'neutouch']

            if feature != 'kernel': line1 = ax.plot(biotac_means['frequency'], biotac_means['mean'], '--x', label=f'BioTAC PAC {tool_length}cm')
            line2 = ax.plot(neutouch_means['frequency'], neutouch_means['mean'], '-o', label=f'NeuTouch {tool_length}cm')

        ax.set_ylabel('MAE (cm)')
        ax.set_xlabel('Frequency (Hz)')
        ax.set_xticks([5, 50, 100, 250,	500,	1000,	2000,	2200, 3000, 4000])
        ax.set_xticklabels([5, '', '', 250,	500,	1000,	2000,	2200, 3000, 4000], rotation='vertical')
        ax.legend()

        fig.tight_layout()
        #plt.savefig(f'question5/{algorithm}_{feature}.pdf')

        plt.show()
        break
    break
```
